```python
def result(home, away, home_score, away_score):
    if (home == 'Barcelona') and (home_score > away_score):
        return "W"
    
    elif (home == 'Barcelona') and (home_score < away_score):
        return "L"
    
    elif (home != 'Barcelona') and (home_score > away_score):
        return "L"
    
    elif (home != 'Barcelona') and (home_score < away_score):
        return "W"
    
    elif (home == 'Barcelona') and (home_score == away_score):
        return "D"
    
    elif (home != 'Barcelona') and (home_score == away_score):
        return "D"
```

```python
barca_fix['result'] = np.vectorize(result)(barca_fix['Home'], barca_fix['Away'], barca_fix['Home_score'], barca_fix['Away_score'])
```

```python
# Barca performances against Real Madrid between 2000-2021 including friendly games
barca_fix[(barca_fix['Home'] == 'Real Madrid') | (barca_fix['Away'] == 'Real Madrid')].sort_values('Season')
```

```python
#  Barca home & away games goals between 2000 - 2021 including friendly games
barca_fix.groupby('Season').sum()
```

```python
# Barca home games winning rate between 2000-2021
plt.figure(figsize=(8,6), dpi=100)
plt.xticks(rotation=90)
sns.countplot(data=barca_fix[barca_fix['Home'] == 'Barcelona'].sort_values('Season') , x='Season', hue='result', palette='magma')
plt.ylabel('Winning Rate');
```

```python
# Barca away games winning rate between 2000-2021
plt.figure(figsize=(8,6), dpi=100)
plt.xticks(rotation=90)
sns.countplot(data=barca_fix[barca_fix['Away'] == 'Barcelona'].sort_values('Season'), x='Season', hue='result', palette='viridis')
plt.ylabel('Winning Rate');
```

```python
barca_fix_d = barca_fix.join(pd.get_dummies(barca_fix['result']))
```

```python
# Numbers of home goals barca scored against opponent between 2000/01 - 2020/21 and their Winning rate
plt.figure(figsize=(30, 15), dpi=300)
plt.xticks(rotation=90)
plt.xscale('linear')
plt.ylabel('Number of goals scored at home')
plt.xlabel('Away Team')
sns.scatterplot(data = barca_fix[barca_fix['Home'] == 'Barcelona'].sort_values('Away') , x='Away', y= 'Home_score', hue='result');
```

```python
# Numbers of away goals barca conceded against opponent between 2000/01 - 2020/21 and their Winning rate
plt.figure(figsize=(30, 15), dpi=300)
plt.xticks(rotation=90)
plt.xscale('linear')
plt.ylabel('Number of goals conceded at away')
plt.xlabel('Home team')
sns.scatterplot(data = barca_fix[barca_fix['Away'] == 'Barcelona'].sort_values('Home') , x='Home', y= 'Home_score', hue='result');
```

```python
# Head to Head Stats of a club against Barca between 2000/01 to 2020/2021
def return_HtHStats(club):
    return barca_fix[(barca_fix['Home'] == club) | (barca_fix['Away'] == club)].sort_values('Season')
```
